```python
import pandas as pd
import numpy as np
```

```python
crime = pd.read_csv('Final_Dataset.csv')
unemployment = pd.read_excel('unemployment.xlsx')
income = pd.read_excel('income.xlsx')
```

```python
unemployment.head()
```

```python
income.head()
```

```python
crime["Unemployment Rate %"] = np.nan

crime["Median Income"] = np.nan

crime.head()
```

```python
#Merge income and crime datasets

for i in range(len(crime)):
    for j in range(len(income)):
        if crime['report_year'][i] == income['Year'][j] and crime['state'][i] == income['State'][j]:
            crime['Median Income'][i]= income['Median Income'][j]
               
```

```python
crime.head()
```

```python
#Merge Unemployment and crime datasets

for i in range(len(crime)):
    for j in range(len(income)):
        if crime['report_year'][i] == unemployment['Year'][j] and crime['state'][i] == unemployment['State'][j]:
            crime['Unemployment Rate %'][i]= unemployment['Unemployment Rate %'][j]
```

```python
crime.head()
```

```python
crime.to_csv('Updated Dataset-Crime, Rent, Income, Unemployment.csv')
```
